```python
from tavily import TavilyClient

class WebResearchAgent:
    def __init__(self, topic):
        self.topic = topic
        self.questions = []
        self.research_data = {}
        self.model = genai.GenerativeModel(model_name="models/gemini-1.5-flash-latest")
        self.tavily = TavilyClient()

    def generate_questions(self):
        prompt = f"Generate 5 to 6 insightful and well-structured research questions about the topic: {self.topic}."
        response = self.model.generate_content(prompt)
        self.questions = [
            q.strip("1234567890). ").strip()
            for q in response.text.strip().split("\n")
            if q.strip()
        ]

        # üëá Print the generated questions
        print("\nüìã Generated Research Questions:")
        for i, q in enumerate(self.questions, 1):
            print(f"{i}. {q}")

    def search_web(self):
        for question in self.questions:
            safe_query = question[:400]
            try:
                result = self.tavily.search(query=safe_query, search_depth="advanced", include_answer=True)
                self.research_data[question] = [
                    {"title": r.get("title"), "content": r.get("content")}
                    for r in result.get("results", [])
                ]
            except Exception as e:
                self.research_data[question] = [{"title": "Search Error", "content": str(e)}]


    def generate_report(self):
        report = f"# üßæ Research Report on **{self.topic}**\n\n"
        report += "## üß† Introduction\nThis report is generated by an AI agent using the ReAct pattern (Reasoning + Acting).\n\n"
        for question in self.questions:
            report += f"### ‚ùì {question}\n"
            if self.research_data.get(question):
                for item in self.research_data[question]:
                    report += f"- **{item['title']}**: {item['content']}\n"
            else:
                report += "- No relevant results found.\n"
            report += "\n"
        report += "## ‚úÖ Conclusion\nThis concludes the structured research conducted by the AI agent.\n"
        return report
```

```python
# Example Topic
topic = input('Enter the topic you want to explore:')

agent = WebResearchAgent(topic)
agent.generate_questions()
agent.search_web()
final_report = agent.generate_report()
```
