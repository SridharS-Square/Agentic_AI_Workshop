```
pdf_files = ["paper1.pdf", "paper2.pdf", "paper3.pdf"]
```

```
from langchain_community.document_loaders import PyPDFLoader

all_docs = []
for pdf in pdf_files:
    loader = PyPDFLoader(f'./paper/{pdf}')
    pages = loader.load()
    all_docs.extend(pages)
```

```
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=200)
chunks = text_splitter.split_documents(all_docs)
```

```
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vector_store = FAISS.from_documents(chunks, embedding_model)
```

```
retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 4})
```

```
user_query = input("Enter your question: ")

relevant_docs = retriever.invoke(user_query)

for i, doc in enumerate(relevant_docs):
    print(f"\n--- Chunk {i+1} ---\n")
    print(doc.page_content[:500])
```

```
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load FLAN-T5 base (small, fast, good quality)
model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

rag_pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=200
)


# Combine document chunks as context
context = "\n\n".join([doc.page_content for doc in relevant_docs])
safe_context = context[:1500]  # Adjust length to ~512 tokens

# Prompt with trimmed context
prompt = f"""Answer the question based on the following context:\n\n{safe_context}\n\nQuestion: {user_query}\nAnswer:"""
response = rag_pipeline(prompt)[0]['generated_text']

print("\n🧠 Final Answer:\n", response)

```

```
# Store conversation history per user
chat_histories = {}

def get_user_history(user_id):
    return chat_histories.get(user_id, [])

def update_user_history(user_id, question, answer):
    history = chat_histories.get(user_id, [])
    history.append((question, answer))
    chat_histories[user_id] = history

def build_prompt_with_history(user_id, user_query, relevant_docs):
    history = get_user_history(user_id)
    history_prompt = "\n".join([f"Q: {q}\nA: {a}" for q, a in history[-3:]])  # Last 3 QAs

    sources = "\n\n".join([f"Chunk {i+1}: {doc.page_content}" for i, doc in enumerate(relevant_docs)])

    prompt = f"""You are a helpful assistant. Here is the conversation so far:\n{history_prompt}\n\nNow answer the user's new question using the following chunks:\n\n{sources}\n\nQ: {user_query}\nA:"""

    return prompt

def ask_question(user_id, user_query):
    # Step 1: Retrieve documents
    relevant_docs = retriever.get_relevant_documents(user_query)

    # Step 2: Build prompt with history
    prompt = build_prompt_with_history(user_id, user_query, relevant_docs)

    # Step 3: Generate answer
    response = rag_pipeline(prompt)[0]['generated_text']

    # Step 4: Update user history
    update_user_history(user_id, user_query, response)

    return response, relevant_docs


    # Simulate chat from 2 users
    users = ["sridhar", "ravi"]

    # Sridhar asks first
    q1 = "What is a retriever in a RAG system?"
    a1, _ = ask_question("sridhar", q1)
    print("👤 Sridhar:\n", a1)

    # Ravi asks different question
    q2 = "What does generator do in RAG?"
    a2, _ = ask_question("ravi", q2)
    print("👤 Ravi:\n", a2)

    # Sridhar asks follow-up
    q3 = "Can you explain with an example?"
    a3, _ = ask_question("sridhar", q3)
    print("👤 Sridhar:\n", a3)

```

```
# Sample evaluation questions
sample_questions = [
    "What are the two main components of the RAG model?",
    "What is the purpose of FAISS in this system?",
    "How does RAG perform compared to other models?",
    "Who supported the research work in this paper?"
]

# Run evaluation loop
for q in sample_questions:
    chat_history = chat_histories.get("eval_user", [])

    # Get relevant documents
    retrieved_docs = retriever.get_relevant_documents(q)
    context = "\n".join([doc.page_content for doc in retrieved_docs])

    # Build chat prompt
    chat_prompt = "\n".join([f"User: {u}\nAI: {a}" for u, a in chat_history])
    full_prompt = f"{chat_prompt}\nUser: {q}\nContext:\n{context}\nAI:"

    # Use the correct pipeline (assuming it's named `pipe`)
    response = rag_pipeline(full_prompt)

    print(f"\n🧠 Q: {q}\n💬 A: {response[0]['generated_text'].strip()}")

    # Save history
    chat_histories["eval_user"] = chat_history + [(q, response[0]['generated_text'].strip())]
```
